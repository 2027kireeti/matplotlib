# using title labels

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.linspace(0,10,10)
y=np.linspace(0,10,10)
plt.plot(x,y)
plt.title("kireeti 1st plot")
plt.xlabel("x-->axis")
plt.ylabel("y-->axis")
plt.show()
```

# using marker

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.linspace(0,10,10)
y=np.linspace(0,10,10)
plt.plot(x,y,marker="o")
plt.title("kireeti 1st plot")
plt.xlabel("x-->axis")
plt.ylabel("y-->axis")
plt.show()
```

```python
plt.figure(figsize=(4,4),num=1)
plt.plot(x,y,marker="o")
plt.title("kireeti ")
plt.xlabel("x-->axis")
plt.ylabel("y-->axis")
plt.show()
```

```python
plt.plot(x,y,marker="o")
plt.title("third plot")
plt.savefig("thord plot")
plt.close()
```
